```python
b2['dataset_size_gb'] = 1
b1['concurrent_level'] = 1
b1 = b1[b2.columns]
```

```python
bb = b1.merge(b2, how='outer', on=['concurrent_level', 'dataset_size_gb', 'cluster_mem_gb'])
bb
```

```python
grbb = bb.groupby(['concurrent_level', 'dataset_size_gb', 'cluster_mem_gb'])
means = grbb.mean()
variances = grbb.var()
means
```

```python
variances
```

```python
collected_data = means[['running_time_x', 'running_time_y']].reset_index()
collected_data['running_time_mean'] = collected_data.apply(lambda x: x['running_time_y']
                                                           if not np.isnan(x['running_time_y']) 
                                                           else x['running_time_x'], axis=1)
collected_data['running_time_var'] = variances.reset_index().apply(lambda x: x['running_time_y'] 
                                                    if not np.isnan(x['running_time_y']) 
                                                    else x['running_time_x'], axis=1)
collected_data = collected_data[['concurrent_level', 'cluster_mem_gb', 'dataset_size_gb', 
                                 'running_time_mean', 'running_time_var']]
collected_data
```

```python
f, axarr = plt.subplots(3)
concurrent_vs_time = collected_data.groupby('concurrent_level').mean().reset_index()
axarr[0].bar(concurrent_vs_time.concurrent_level.tolist(), concurrent_vs_time.running_time_mean.tolist())
axarr[0].set_title('concurrent_level')

mem_vs_time = collected_data.groupby('cluster_mem_gb').mean().reset_index()
axarr[1].bar(mem_vs_time.cluster_mem_gb.tolist(), mem_vs_time.running_time_mean.tolist())
axarr[1].set_title('cluster_mem_gb')

data_vs_time = collected_data.groupby('dataset_size_gb').mean().reset_index()
axarr[2].bar(data_vs_time.dataset_size_gb.tolist(), data_vs_time.running_time_mean.tolist())
axarr[2].set_title('dataset_size_gb')

plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix
collected_data['running_time_mean_log'] = collected_data.running_time_mean.map(np.log)
scatter_matrix(collected_data, alpha=0.2, figsize=(16, 16), diagonal='kde')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import cross_val_predict

X = collected_data[['concurrent_level', 'cluster_mem_gb', 'dataset_size_gb']].values
y = collected_data.running_time_mean.values
lr = LinearRegression(fit_intercept=True, normalize=False)
# predicted = cross_val_predict(lr, X, y, cv=10)
lr.fit(X, y)
predicted = lr.predict(X)

print lr.score(X, y)
fig, ax = plt.subplots()
ax.scatter(y, predicted)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
print lr.coef_, lr.intercept_
```

```python
# Predict the running time in log domain gives better R2 score

X = collected_data[['concurrent_level', 'cluster_mem_gb', 'dataset_size_gb']].values
y_log = collected_data.running_time_mean_log.values
lr_log = LinearRegression(fit_intercept=True, normalize=False)
# predicted = cross_val_predict(lr, X, y, cv=10)
lr_log.fit(X, y_log)
predicted = lr_log.predict(X)

print lr_log.score(X, y_log)
fig, ax = plt.subplots()
ax.scatter(y_log, predicted)
ax.plot([y_log.min(), y_log.max()], [y_log.min(), y_log.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
print lr_log.coef_, lr_log.intercept_
```

```python
def compute_running_time_exp(x):
    return np.exp(0.024 * x['concurrent_level'] - 0.00119 * x['cluster_mem_gb'] + 
                  0.023137 * x['dataset_size_gb'] + 3.008179)

def compute_running_time(x):
    return 70 - 0.5121 * x['concurrent_level'] - 0.21307 * x['cluster_mem_gb'] + 1.99968 * x['dataset_size_gb']

def compute_running_time_binhle(x):
    return 3 * x['concurrent_level'] * x['dataset_size_gb'] / (x['cluster_mem_gb'] / 32)

collected_data['predicted_running_time'] = collected_data.apply(compute_running_time, axis=1)
collected_data['predicted_running_time_exp'] = collected_data.apply(compute_running_time_exp, axis=1)
collected_data['predicted_running_time_binhle'] = collected_data.apply(compute_running_time_binhle, axis=1)

for c in ['predicted_running_time', 'predicted_running_time_exp', 'predicted_running_time_binhle']:
    mse = np.sqrt((np.sum(np.power(collected_data[c] - collected_data.running_time_mean, 2)) 
    / len(collected_data)))
    
    print 'RMSE for {}: {}'.format(c, mse)
    
collected_data[['concurrent_level', 'cluster_mem_gb', 'dataset_size_gb', 'running_time_mean',
                'predicted_running_time', 'predicted_running_time_exp', 'predicted_running_time_binhle']]
```

```python
# Wayyyyy overkil here

from keras.layers.advanced_activations import PReLU
from keras.layers.core import Dense, Dropout, Activation
from keras.layers.normalization import BatchNormalization
from keras.models import Sequential

hidden_units = 1000
output_units = 1
batch_size = 1

model = Sequential()
model.add(Dense(input_dim=X.shape[1], output_dim=hidden_units, init='glorot_normal'))
# model.add(PReLU(input_shape=(hidden_units,)))
model.add(Activation('relu'))
# model.add(Dropout(input_dropout))

model.add(Dense(input_dim=hidden_units, output_dim=hidden_units, init='glorot_normal'))
# model.add(PReLU(input_shape=(hidden_units,)))
model.add(Activation('relu'))
# model.add(BatchNormalization(input_shape=(hidden_units,)))
# model.add(Dropout(hidden_dropout))

model.add(Dense(input_dim=hidden_units, output_dim=hidden_units, init='glorot_normal'))
# model.add(PReLU(input_shape=(hidden_units,)))
model.add(Activation('relu'))
# model.add(BatchNormalization(input_shape=(hidden_units,)))

model.add(Dense(input_dim=hidden_units, output_dim=output_units, init='glorot_normal'))
model.add(Activation('linear'))
model.compile(loss='mean_squared_error', optimizer='rmsprop')

X = collected_data[['concurrent_level', 'cluster_mem_gb', 'dataset_size_gb']].values
mean_X = X.mean(0)
stdev_X = np.sqrt(X.var(0))
X = (X - mean_X) / stdev_X
y = collected_data.running_time_mean.values
model.fit(X, y, nb_epoch=50, batch_size=batch_size, verbose=5)
predict_nn = model.predict(X)

collected_data['predicted_running_time_nn'] = predict_nn
mse = np.sqrt((np.sum(np.power(collected_data.predicted_running_time_nn - collected_data.running_time_mean, 2)) 
    / len(collected_data)))
print 'RMSE for neural net: {}'.format(mse)

fig, ax = plt.subplots()
ax.scatter(y, predict_nn)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
